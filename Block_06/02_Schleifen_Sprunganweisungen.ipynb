# Schleifen
Immer, wenn wir bestimmte Programmteile wiederholen möchten,können wir die Schleife als iterative Grundstruktur nutzen.

## for-Schleife
Die erste Schleifenart ist die for-Schleife.<br>
Sie beinhaltet zur Schleifensteuerung:
<ol style="list-style-type:numeric">
    <li><b>Startwert:</b> Mit diesem Wert beginnt die for-Schleife</li>
 <li><b>Schrittweite:</b> Nach jedem Durchlauf wird die Schrittweite auf die <b>Laufvariable</b> addiert</li>
 <li><b>Endwert:</b> Bei jedem Schleifendurchlauf wird geprüft, ob der Endwert erreicht ist, solange dies noch nicht der Fall ist, wird der Schleifenrumpf weiter durchlaufen</li>
     </ol> 
 <b>Hinweis: </b> Startwert und Schrittweite sind optional. Werden diese weggelassen, geht Python von dem    Startwert 0 und der Schrittweite 1 aus.





```python
zähler = 1
for i in range (10, 0, -2):
    # Die Schrittweite ist negativ, sodass die Schleife
    # bei 10 anfängt und bei dem Durchlauf -2 auf die
    # Laufvariable i rechnet
    print("Im", i, ". Durchlauf der Schleife hat die Laufvariable den Wert", i)
    zähler += 1
```

## while-Schleife

Die zweite Schleifenart ist die while-Schleife.
Sie beinhaltet zur Schleifensteuerung eine BEdingung und wird deshlab auch bedingte Schleife genannt.

```python
pin = 9384
pin_eingabe = 0

while (pin_eingabe != pin):
    # es wird solange nach der Pin-Eingabe gefragt,
    # bis dieser richtig eingegeben wurde
    pin_eingabe = int(input("Bitte Pin eingeben"))
    
print("PIN stimmt")
```

### !!!Achtung!!!
Es lassen sich auch sehr schnell Endlosschleifen erzeugen. Das kann manchmal gewollt sein, meist jedoch führt diese zu unerwünschtem Programmverhalten.

```python
a = 1
while True:
    a *= 2
    print(a)
```

### Abbruch von Schleifen, Springen aus Schleifen
Auch innerhalb von Schleifen können bedingte Anweisungen stehen und es ist möglich bei bestimmten Bedingungen aus einer Schleife herauszuspringen oder diese Schelife sofort mit der nächsten Iteration weiter laufen zu lassen.<br>
Dazu werden die Schlüsselwörter <b>break</b> und <b>continue</b> genutzt.<br>
<b>break</b> bricht die aktuelle Schleife sofrt ab und springt an das Ende der Einrückung.<br>
<b>continue</b> sorgt für einen Sprung direkt zurück in Zeile der Schleifensteuerung.
